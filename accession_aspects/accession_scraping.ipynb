```python
async def fetch_all_accessions(url, max_pages=20):
    browser = await launch(args=['--no-sandbox', '--disable-setuid-sandbox', '--disable-infobars'])
    page = await browser.newPage()
    await page.goto(url, waitUntil='networkidle0')  # Wait for page to load

    all_accessions = set()

    current_page = 1
    while current_page <= max_pages:
        await asyncio.sleep(5)  # Adjust sleep time as necessary

        html = await page.content()
        soup = BeautifulSoup(html, 'html.parser')
        dd_elements = soup.find_all('dd')
        for dd in dd_elements:
            all_accessions.add(dd.get_text(strip=True))

        # Check if the Next button is active
        active_next_button_selector = 'a.active.page_link.next'  # Update this selector for an active Next button
        inactive_next_button_selector = 'span.inactive.page_link.next'  # Selector for an inactive Next button

        active_next_button = await page.querySelector(active_next_button_selector)
        inactive_next_button = await page.querySelector(inactive_next_button_selector)

        if active_next_button:
            # If an active Next button is found, click it to load the next page
            await active_next_button.click()
            current_page += 1
        elif inactive_next_button:
            # If an inactive Next button is found, stop the loop
            print("No more pages. Stopping.")
            break
        else:
            # If neither active nor inactive Next button is found, perhaps due to an error or unexpected page structure
            print("Next button not found. Please check the selectors.")
            break

    await browser.close()
    return sorted(all_accessions)
```

```python
all_accessions_global = None

async def main():
    global all_accessions_global
    start_url = 'https://www.ncbi.nlm.nih.gov/sra?linkname=bioproject_sra_all&from_uid=237362'
    all_accessions = await fetch_all_accessions(start_url)  # Ensure await is used here

    print("Fetched accession numbers:")
    print(len(all_accessions))
    for accession in all_accessions:
        print(accession)

    all_accessions_global = all_accessions

# In a Jupyter Notebook cell, you would call your main function like this:
await main()
```

```python
def format_accession_urls(accession_numbers):
    base_url = 'https://www.ncbi.nlm.nih.gov/sra/{}[accn]'
    urls_list = []  # Renamed from 'list' to 'urls_list' for clarity

    for accession in accession_numbers:  # Assuming this is the correct list you intended to iterate over
        url = base_url.format(accession)
        response = requests.get(url)
        html_content = response.text
        soup = BeautifulSoup(html_content, 'html.parser')
        pattern = re.compile(r'SRR\d+')

        for a_tag in soup.find_all('a', href=True):
            href = a_tag['href']
            if pattern.search(href):
                # Assuming the direct download URL can be derived by modifying the href
                # For example, appending "&display=download" if that's how NCBI's URLs work
                # Adjust this line according to the actual pattern needed to navigate to the downloads section
                full_url = 'https:' + href if href.startswith('//') else href
                download_url = full_url + "&display=download"  # This is an example adjustment
                print("Download URL:", download_url)
                urls_list.append(download_url)

    return urls_list


download_urls = format_accession_urls(all_accessions_global)

import re

def parse_srr_accessions(url_list):
    """
    Parses SRR accession numbers from a list of URLs.

    Parameters:
    url_list (list): A list of URLs as strings.

    Returns:
    list: A list of extracted SRR accession numbers.
    """
    srr_pattern = re.compile(r'SRR\d+')
    accessions = []

    for url in url_list:
        found = srr_pattern.search(url)
        if found:
            accessions.append(found.group())

    return accessions

acc = parse_srr_accessions(download_urls)
for item in acc:
  print(item)
```
